```python
# تثبيت المكتبات اللازمة
!pip install pyspark
!pip install rasterio
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# ملاحظة: استبدل الأسماء بأسماء الملفات اللي عندك بعد رفعها
# عادة Band 1 هو الأحمر و Band 2 هو الأشعة تحت الحمراء في مشروعك
try:
    with rasterio.open('band1.tif') as red_dataset:
        red_band = red_dataset.read(1).astype('float32')

    with rasterio.open('band2.tif') as nir_dataset:
        nir_band = nir_dataset.read(1).astype('float32')

    # معادلة NDVI مع إضافة رقم صغير جداً لمنع القسمة على صفر
    ndvi = (nir_band - red_band) / (nir_band + red_band + 1e-10)

    # عرض النتيجة
    plt.figure(figsize=(10, 10))
    plt.imshow(ndvi, cmap='RdYlGn') # خريطة ألوان من الأحمر للأخضر
    plt.colorbar(label='NDVI Index')
    plt.title('Vegetation Health Map (NDVI)')
    plt.show()

except Exception as e:
    print(f"Error: {e} - تأكد من وجود ملفات الـ TIF في جانب الشاشة")
```
